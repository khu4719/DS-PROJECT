```python
# DATA ANALYSIS PROJECT ON WORLD HAPPINESS REPORT 2019

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df=pd.read_csv(r"C:\Users\Student\Downloads\2019.csv")
df.isnull().sum()
df.dropna(inplace=True)

# Top 10 Happiest Countries
top_10=df.sort_values(by="Score",ascending=False).head(10)
top_10[["Country or region","Score"]] 

# # Bottom 10 least Happy Countries
bottom_10=df.sort_values(by="Score",ascending=True).head(10)
bottom_10[["Country or region","Score"]] 

# # Bar Chart
plt.figure(figsize=(10,6))
plt.barh(top_10['Country or region'], top_10['Score'], color='green')
plt.xlabel('Score')
plt.title('Top 10 Happiest Countries')
plt.gca().invert_yaxis()
plt.show()

# Pie Chart 
country = df[df['Country or region'] == 'Finland'].iloc[0]
labels = ['Economy', 'Health', 'Freedom', 'Trust', 'Generosity']
sizes = [
    country['GDP per capita'],
    country['Healthy life expectancy'],
    country['Freedom to make life choices'],
    country['Perceptions of corruption'],
    country['Generosity']
]

plt.figure(figsize=(7,7))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title('Happiness Factors Contribution - Finland')
plt.show()


# Plot marker graph
plt.figure(figsize=(12, 6))
plt.plot(top_10['Country or region'], top_10['Score'], 
         marker='o', linestyle='-', color='purple')

plt.title(f'Top 10 Countries by Happiness Score', fontsize=14)
plt.xlabel('Country')
plt.ylabel('Happiness Score')
plt.xticks(rotation=45, ha='right')
plt.grid(True)
plt.tight_layout()
plt.show()

```
